Exploring and Visualizing a Simple Dataset
This task involves loading the famous Iris dataset and performing basic exploratory data analysis (EDA) and visualizations to better understand the data.

```python
#import libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
print("Shape of dataset:", df.shape)
```

```python
print("\nColumns in dataset:", df.columns)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#checking for missing values
df.isnull().sum()
```

```python
#Scatter plot to analyze relationships between variables.
sns.scatterplot(data=df, x= 'sepal_length', y= 'sepal_width', hue = 'species')
plt.title('Sepal Length vs Width')
plt.show()
```

```python
sns.scatterplot(data=df, x= 'petal_length', y= 'petal_width', hue = 'species')
plt.title('Petal Length vs Width')
plt.show()
```

```python
#Histogram to examine data distribution.
df.hist(figsize=(10, 8), color = 'lightblue',edgecolor ='black')
plt.tight_layout()
plt.show()
```

```python
# Box plot to detect outliers and spread of values.
plt.figure(figsize =(10,8))
sns.boxplot(data=df)
plt.title('Box plot of all numerical features')
plt.show()
```
